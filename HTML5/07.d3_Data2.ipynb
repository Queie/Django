### <strong>01 셀렉션과 바인딩
1. <strong>.select(), .selectAll()</strong> : DOM 요소선택, 본문에 요소가 없을땐 <strong>.enter()</strong>로 추가
1. <strong>.data()</strong> : DOM 요소에 배열을 연결
1. <strong>.enter()</strong> : <strong>.enter()</strong>는 DOM 보다 데이터가 더 많을때 공간을 넓히고, 
1. <strong>.enter()</strong> : <strong>.exit()</strong>는 데이터가 DOM 보다 적을 때 호출을 종료한다
1. <strong>.append(), .insert()</strong> : .append() 요소를 추가, .insert()는 요소추가 위치를 지정가능
1. <strong>.attr()</strong> : .append()추가 태그에 (class='cities') 속성이 자동적 추가된다
1. <strong>.html()</strong> : 전통적 DOM 요소의 콘텐츠 내용을 설정

```python
%%d3
<g id='tags'></g>
<script>
d3.csv("./data/cities.csv",
       function(error,data) { dataViz(data) });
       function dataViz(incomingData) {
            d3.select("#tags").selectAll("p.cities")   // g_2 태그 내부에 p 태그 없어서 '빈 셀렉션'선택
              .data(incomingData)                    // 데이터를 셀렉션에 binding
              .enter()                               // 셀렉션 DOM 보다 데이터가 더 많을 때
              .append("p")                           // 셀렉션 에 p tag 를 추가
              .attr("class","cities")                // 생성요소의 클래스 설정
              .html(function(d,i) {return d.label})} // p 의 입력 내용을 설정
</script>
```

```python
%%d3  
<!-- Json 객체에서 d3.nest() 로 배열 객체를 활용한 '막대그래프' -->
<svg id='csv2'>
</svg>
<script>
d3.json("./data/tweets.json", function(error,data) {dataViz(data.tweets)});

function dataViz(incomingData) {
    var nestedTweets = d3.nest()
        .key(function (el) {return el.user;})
        .entries(incomingData);

    nestedTweets.forEach(function (el) { // Object : nestedTweets[0,1,2 ...] Array 객체 
        el.numTweets = el.values.length; // Key:.numTweets, Value:.values.length
    })

    var maxTweets = d3.max(nestedTweets, function(el) {return el.numTweets;});
    var yScale    = d3.scale.linear().domain([0,maxTweets]).range([0,100]);
          d3.select("#csv2")
            .selectAll("rect")
            .data(nestedTweets)
            .enter()
            .append("rect")
            .attr("width", 50)
            .attr("height", function(d) {return yScale(d.numTweets);})
            .attr("x", function(d,i) {return i * 60;})
            .attr("y", function(d) {return 100 - yScale(d.numTweets);})
            .style("fill", "blue")
            .style("stroke", "red")
            .style("stroke-width", "1px").style("opacity", .25);
            }</script>
```

### <strong>02 채널의 설정
객체에서 다변량 데이터를 호출하는 경우

```python
%%d3
<!-- 산포도를 그리는 경우 -->
<svg></svg>
<script>
d3.json("./data/tweets.json", 
        function(error,data) {dataViz(data.tweets)});  // dataViz 함수로 연산한 결과를 출력

function dataViz(Json) {Json.forEach(function (el) {   // dataViz 함수를 정의한다
    el.impact = el.favorites.length + el.retweets.length;  // .favorites 과 .retweets 를 합산한 영향력 점수를 생성
    el.tweetTime = new Date(el.timestamp);                 // ISO 8906 데이터를 Date 형태로 변환
    })
    // Json 에서 필요한 value들을 추출한다
    var maxImpact = d3.max(Json, function(el) {return el.impact;});
    var startEnd  = d3.extent(Json, function(el) {return el.tweetTime;});
    var timeRamp = d3.time.scale().domain(startEnd).range([20,480]);     // startEnd 시간 배열값을 re-scaleing
    var yScale = d3.scale.linear().domain([0,maxImpact]).range([0,460]); 
    var radiusScale = d3.scale.linear().domain([0,maxImpact]).range([1,20]);  
    var colorScale  = d3.scale.linear().domain([0,maxImpact]).range(["white","#990000"]);
    d3.select("svg")             // visual 연산을 시작
      .selectAll("circle")
      .data(Json)
      .enter() 
      .append("circle")
      .attr("r",  function(d)    {return radiusScale(d.impact);})
      .attr("cx", function(d,i)  {return timeRamp(d.tweetTime);})
      .attr("cy", function(d)    {return 480 - yScale(d.impact);})
      .style("fill", function(d) {return colorScale(d.impact);})
      .style("stroke", "black")
      .style("stroke-width", "1px");
};</script>
```

### <strong>02 기타 메서드들의 역활
데이터 바인딩 후처리
1. <strong>.enter()</strong>   : 바인딩 데이터가 더 많을 때 공간을 넓힌다
1. <strong>.update()</strong>  : 그림요소를 생성한 함수를 다시 적용한다
1. <strong>.exit()</strong>    : 바인딩 데이터가 적을 때 작업을 종료한다

```python
%%d3
<!-- 산포도 위에 text 덧붙이기 -->
<svg></svg>
<script>
d3.json("./data/tweets.json", 
        function(error,data) {dataViz(data.tweets)});  // dataViz 함수로 연산한 결과를 출력

function dataViz(Json) {Json.forEach(function (el) {   // dataViz 함수를 정의한다
    el.impact = el.favorites.length + el.retweets.length;
    el.tweetTime = new Date(el.timestamp);
    })

    var maxImpact = d3.max(Json, function(el) {return el.impact;});   // value들을 추출한다
    var startEnd  = d3.extent(Json, function(el) {return el.tweetTime;});
    var timeRamp = d3.time.scale().domain(startEnd).range([20,480]);
    var yScale = d3.scale.linear().domain([0,maxImpact]).range([0,460]); 
    var radiusScale = d3.scale.linear().domain([0,maxImpact]).range([1,20]);  
    var colorScale  = d3.scale.linear().domain([0,maxImpact]).range(["white","#990000"]);
    d3.select("svg")
      .selectAll("circle")
      .data(Json)
      .enter() 
      .append("circle")
      .attr("r",  function(d)    {return radiusScale(d.impact);})
      .attr("cx", function(d,i)  {return timeRamp(d.tweetTime);})
      .attr("cy", function(d)    {return 480 - yScale(d.impact);})
      .style("fill", function(d) {return colorScale(d.impact);})
      .style("stroke", "black")
      .style("stroke-width", "1px");
                                    // 부모의 데이터 객체를 상속하므로 동일한 데이터의 가공이 가능
    var tweetG = d3.select("svg")   // tweetG 함수에서 태그와 텍스트를 자동으로 추가한다
                    .selectAll("g") // g 태그 : 구조화된 문자요소와 결합한다
                    .data(Json)
                    .enter()
                    .append("g")
                    .attr("transform", function(d) {
                            return "translate("+ timeRamp(d.tweetTime) + 
                                            "," + (480-yScale(d.impact))+ ")"})
        tweetG.append("circle")
                .attr("r", function(d) {return radiusScale(d.impact)})
                .style("fill", "#990000")
                .style("stroke", "black")
                .style("stroke-width", "1px")
        tweetG.append("text")                 // text를 읽기좋게 .getHours() 를 사용한다
                .text(function(d) {return d.user + "-" + d.tweetTime.getHours()});}
</script> 
```
